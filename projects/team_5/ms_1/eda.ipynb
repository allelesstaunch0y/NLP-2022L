```python
import torchtext.datasets as datasets
from itertools import chain
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

data = datasets.AG_NEWS()
```

```python
def prepare_hist(data_iterator):
    y = [len(sample[1].split()) for sample in data_iterator]
    plt.hist(y, bins=30)
    plt.yscale('log')
    plt.xlabel('Word count')
    plt.ylabel('Number of samples')
    plt.savefig('word-counts.pdf')
    plt.show()
```

```python
prepare_hist(chain(data[0], data[1]))
```

```python
def prepare_class_hist_alpha(data_iterator):
    y = {}
    for c, text in data_iterator:
        if c not in y:
            y[c] = []
        y[c].append(text)
    for c, data in y.items():
        plt.hist([len(x.split()) for x in data], bins=30, alpha=0.5)
        plt.yscale('log')
        plt.xlabel('Word count')
        plt.ylabel('Number of samples')

    plt.savefig('word-counts-within-classes-overlay.pdf')

    plt.show()
```

```python
prepare_class_hist_alpha(chain(data[0], data[1]))
```

```python
def prepare_class_hist(data_iterator):
    y = {}
    matplotlib.rcParams.update({'font.size': 22})
    for c, text in data_iterator:
        if c not in y:
            y[c] = []
        y[c].append(text)
    for c, data in y.items():
        plt.hist([len(x.split()) for x in data], bins=30)
        plt.yscale('log')
        plt.xlim([0,2500])
        plt.ylim([0,1e5])
        plt.savefig(f'word-counts-class{c}.pdf')
        plt.show()
```

```python
prepare_class_hist(chain(data[0], data[1]))
```

```python
data = datasets.IMDB()
prepare_hist(chain(data[0], data[1]))
prepare_class_hist_alpha(chain(data[0], data[1]))
prepare_class_hist(chain(data[0], data[1]))
```
